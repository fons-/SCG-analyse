# Een paar gedachten over het clusterprobleem

**Het clusterprobleem in begrijpelijke taal**

Zij $X \subseteq \mathbb{R}^{3}$ de verzameling ontladingen in een kabel, waarbij we de assen identificeren met de locatie, tijd en ladingsgrootte. $X$ is natuurlijk eindig. Zij $S \subseteq \mathbb{R}^{2}$, en zij $\pi:\mathbb{R}^{3}\to\mathbb{R}^{2}$ met $\pi(s,t,c)=(s,t)$ de projectie naar $\mathbb{R}^{2}$, dan kunnen we de *dichtheid* $\rho(S)$ van een verzameling logischerwijze definiëren door: $\rho(S)=\frac{\#\{x\in X|\pi(x)\in S\}}{A(S)}$ waarbij A de gebruikelijke oppervlakte geeft van een verzameling. Als we de ladingsgrootte ook mee willen rekenen, en we zeggen dat $\pi_{C}(s,t,c)=c$, dan kunnen we $\rho_{C}(S)=\frac{\sum_{y\in\{x\in X|\pi(x)\in S\}}\pi_{C}(y)}{A(S)}$ definiëren als de *gewogen dichtheid* van een deelverzameling van $\mathbb{R}^{2}$. 

Het doel is om samenhangende $S_{i}\subseteq\mathbb{R}^{2}$ te vinden met een voldoende grote oppervlakte (anders worden bijvoorbeeld drie punten die toevallig bij elkaar liggen onterecht aangemerkt als cluster) zodanig dat alle $\rho(S_{i})$ of $\rho_{C}(S_{i})$ (of beide) duidelijk stukken groter zijn dan andere "willekeurige" $\rho(R)$ of $\rho_{C}(R)$ met $R\cap S_{i}=\varnothing$ voor alle $i$. Deze $R$ kunnen we natuurlijk niet echt willekeurig kiezen. Als $R=\mathbb{R}^{2}\setminus\bigcup_{i}S_{i}$ bijvoorbeeld, dan $\rho(R)=0$ en als $R={x}$ met $x\in X\setminus\bigcup_{i}S_{i}$ dan is $\rho(R)=\infty$, ongeacht wat de $S_{i}$'s zijn. We moeten dus voorwaarden stellen aan de verzameling $R$

Gezien $X$ eindig is kunnen we minima en maxima nemen van elementen van $X$. We weten dat $X\subseteq[\min{\pi_{S}(X)},\max{\pi_{S}(X)}]\times[\min{\pi_{T}(X)},\max{\pi_{T}(X)}]\times\mathbb{R}:=W$ waarbij $\pi_{S}$ en $\pi_{T}$ de projecties naar respectievelijk de plaats en de tijd zijn. We kunnen voor $R$, dus de verzameling die *geen* niet-lege doorsnede met een cluster mag hebben, eisen dat deze in $\pi(W)$ moet zitten. Net als voor de verzamelingen $S_{i}$ kunnen we eisen dat $R$ een voldoende grote oppervlakte heeft, om te voorkomen dat de dichtheid van $R$ arbitrair hoog kan worden.

**Een mogelijke definitie van een cluster**

Zij $X \subseteq \mathbb{R}^{3}$ de verzameling ontladingen in een kabel, en $W:=[\min{\pi_{S}(X)},\max{\pi_{S}(X)}]\times[\min{\pi_{T}(X)},\max{\pi_{T}(X)}]\times\mathbb{R}$. Dan heet $S=\{S_{1},S_{2},...,S_{n}\}\subseteq\{0,1,...,n\}\times\mathbb{R}^{2}$ een *verzameling clusters met afstand minstens $r_{s}$ en tijd minstens $r_{t}$* als:

-$S_{i}=\bigcup_{j}[a_{j},b_{j}]\times[c_{j},d_{j}]\subseteq\mathbb{R}^{2}$ met $b_{j}-a_{j}>r_{s},d_{j}-c_{j}>r_{t}$ voor alle $1\leq i \leq n$

-$|\pi_{S}(x)-\pi_{S}(y)|>r_{s}$ of $|\pi_{T}(x)-\pi_{T}(y)|>r_{t}$ voor alle $x\in S_{i}$,$y\in S_{j}$, voor alle paren $1\leq i\neq j\leq n$.

-Als $R=\bigcup_{j}[a_{j},b_{j}]\times[c_{j},d_{j}]\subseteq\pi(W)\setminus\bigcup_{i}S_{i}$ met $b_{j}-a_{j}>r_{s},d_{j}-c_{j}>r_{t}$, dan $\rho_{C}(R)<2*\min_{i}{\rho_{C}(S_{i})}$ voor alle $1\leq i\leq n$

**Werkt deze definitie?**

Deze definitie geeft alleen maar aan wanneer een bepaalde verzameling deelverzamelingen van $\mathbb{R}^{2}$ een clusterverzameling is. Dit geeft dus alleen een manier om te controleren of een algoritme het gewenste resultaat heeft. Men kan direct opmerken dat de definitie het toelaat dat $S=\{\pi(W)\}$, dus er zijn vrijwel altijd meerdere mogelijkheden voor wat $S$ kan zijn.

De eerste voorwaarde geeft aan dat het cluster niet te klein mag zijn. Dit is op basis van de getallen $r_{t}$ en $r_{s}$, die wellicht voorspelbaar zijn. Het getal $r_{s}$ bijvoorbeeld zou als volgt bepaald kunnen worden: stel dat zich op afstand $x$ een zwak punt in de kabel bevindt, dan is de afstand van een ontlading tot $x$ misschien (bij benadering) wel normaal verdeeld met standaardafwijking $\sigma$. Het is dan niet onredelijk om het getal $2\sigma$ te nemen voor $r_{s}$. Mijn eigen verwachting is dat de standaardafwijkingen per zwak punt in de kabel niet erg van elkaar verschillen.

De tweede voorwaarde zorgt ervoor dat twee aparte clusters een minimum afstand tot elkaar hebben. Dit om te zorgen dat twee "clusters" die dicht naast elkaar liggen als één beschouwd kunnen worden.

De derde, tevens ook de belangrijkste voorwaarde stelt dat als we een willekeurig ander gebied pakken van voldoende grootte, dat dit gebied dan minimaal een twee keer zo kleine dichtheid heeft in vergelijking met een willekeurig cluster. De "2" in de definitie is helaas een getal dat ruimte voor discussie overlaat. Dit getal zou kleiner of groter kunnen worden, afhankelijk van de "grens" waarop iets wél als cluster beschouwd kan worden en wanneer niet. Mijn persoonlijke inschatting is dat het getal 2 redelijk werkt.

# Algoritme

Het algoritme dat ik heb bedacht verdeelt de data in discrete "hokjes". Daarna kijkt het algoritme wat het hokje met de grootste dichtheid is. Die dichtheid wordt vervolgens als referentie gebruikt. Als een hokje een dichtheid heeft die groter is dan een getal $c$ maal de maximale dichtheid, dan ziet het algoritme dat hokje als een hokje dat bij een cluster hoort. Vervolgens worden al deze hokjes met een ander algoritme zo geordend dat er groepjes ontstaan, die in het plaatje elk een andere kleur hebben. 

**Basisfuncties en modules inladen**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = [12, 6]
```

```python
def findall(string,char):
    arr=np.array([])
    for i in range(len(string)):
        if string[i]==char:
            arr=np.append(arr,[i])
    return arr
```

```python
def open_file(file):
    lst=np.array(open(file,"r").readlines())
    timearr=np.array([])
    placearr=np.array([])
    chargearr=np.array([])
    for i in range(1,len(lst)):
        if lst[i][19:]!=';;\n':
            fnd=findall(lst[i],";")
            timearr=np.append(timearr,[i])
            placearr=np.append(placearr,[float(lst[i][int((fnd[0]+1)):int(fnd[1])])])
            chargearr=np.append(chargearr,[float(lst[i][int(fnd[1]+1):])])
    return np.vstack((timearr,placearr,chargearr))
```

**Code**

```python
def get_box_y(index,boxnumber,maxtime,data):
    return min(boxnumber-1,int(data[0][index]*boxnumber/maxtime))
```

```python
def get_box_x(index,boxnumber,maxplace,data):
    return min(boxnumber-1,int(data[1][index]*boxnumber/maxplace))
```

```python
def track_groups(elt,todo,ijlist):
    if elt in ijlist:
        todo+=[elt]
        ijlist.remove(elt)
def group_boxes(grid,condition,sizex,sizey):
    ijlist=[]
    for i in range(sizex):
        for j in range(sizey):
            if grid[i,j]>=condition:
                ijlist+=[[i,j]]
    groups=[]
    groupcount=0
    for ij in ijlist:
        groups+=[[ij]]
        groupstart=True
        todo=[ij]
        ijlist.remove(ij)
        while len(todo)>0:
            for p in todo:
                track_groups([p[0]+1,p[1]],todo,ijlist)
                track_groups([p[0]-1,p[1]],todo,ijlist)
                track_groups([p[0],p[1]+1],todo,ijlist)
                track_groups([p[0],p[1]-1],todo,ijlist)
                if groupstart==True:
                    groupstart=False
                else:
                    groups[groupcount]+=[p] 
                todo.remove(p)
        groupcount+=1
    return np.array(groups)
```

```python
%matplotlib notebook

def show_discrete_clusters(file,boxesx,boxesy,sensitivity):

    #met data spelen

    data=open_file(file)
    grid=np.zeros((boxesx,boxesy))
    maxplace=max(data[1])
    maxtime=max(data[0])
    maxcharge=max(data[2])
    sstep=maxplace/boxesx
    tstep=maxtime/boxesy
    datalength=len(data[0])

    #gewoon scatterplot

    plt.figure("plaatjen")
    plt.axis([0,maxplace,0,maxtime])
    plt.xlabel("Plaats")
    plt.ylabel("Tijd")
    plt.scatter(data[1],data[0],marker="o",s=1,c="black")

    #data discretiseren

    for i in range(datalength):
        grid[get_box_x(i,boxesx,maxplace,data)][get_box_y(i,boxesy,maxtime,data)]+=1

    maxdense=max(grid.flatten())

    #data groeperen

    groups=group_boxes(grid,sensitivity*maxdense,boxesx,boxesy)

    colorcycle=["red","green","blue","yellow","orange","purple","pink","cyan","steelblue"]
    for i in range(len(groups)):
        for j in range(len(groups[i])):
            plt.gca().add_patch(Rectangle((groups[i][j][0]*sstep,groups[i][j][1]*tstep),sstep,tstep,linewidth=1,edgecolor=colorcycle[i%len(colorcycle)],facecolor=colorcycle[i%len(colorcycle)],alpha=0.5))

    plt.show()

show_discrete_clusters("/home/geelhoed/GitHub/SCG-analyse/data/origineel/2870-pd.csv",2**8,2**5,0.01)
```

**We moeten dus de verdeling van de hoogste waarden van de dichtheid bepalen**

```python
%matplotlib notebook
def densityplot(file,lb):
    data=open_file(file)
    boxesx=2**8
    boxesy=2**6
    grid=np.zeros((boxesx,boxesy))
    maxplace=max(data[1])
    maxtime=max(data[0])
    datalength=len(data[0])

    #data discretiseren

    for i in range(datalength):
        grid[get_box_x(i,boxesx,maxplace,data)][get_box_y(i,boxesy,maxtime,data)]+=1

    density=np.sort(grid.flatten())
    plt.plot(np.arange(boxesx*boxesy),density,label=lb)

filelist=["1512","2063","2145","2806","2870","2979","2980","3010"]
plt.xlim(16300, 2**8*2**6)
for i in filelist:
    densityplot("/home/geelhoed/GitHub/SCG-analyse/data/origineel/"+i+"-pd.csv",i)
plt.legend(loc=2)
plt.show()
```
